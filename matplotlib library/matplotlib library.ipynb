### matplotlib library
- plotting library

```python
# plot sin(x) and cos(x) in matplotlib with grids and coordinate system x and y axis

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 4 * np.pi, 50)
y = np.sin(x)
z = np.cos(x)

plt.plot(x, y, label='sin(x)')
plt.plot(x, z, label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin(x) and cos(x)')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# linspace(start, stop, span)
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Calculate tan(x) and cot(x)
tan_x = np.tan(x)
cot_x = 1 / np.tan(x)

# plt.figure(figsize=(width, height)) in inches
plt.figure(figsize=(8, 6)) 

# Plot tan(x)
plt.plot(x, tan_x, label='tan(x)', color='b')

# Plot cot(x)
plt.plot(x, cot_x, label='cot(x)', color='r')

# Set the title and labels
plt.title('Plot of tan(x) and cot(x)')
plt.xlabel('x')
plt.ylabel('y')

# Set the x and y axis limits
plt.xlim(-2 * np.pi, 2 * np.pi)
plt.ylim(-1000, 1000)

# Add grid and axis lines
# Make grid lines more dense
plt.minorticks_on()
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.grid(which='minor', linestyle=':', linewidth=0.5)

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for x
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Calculate tan(x) and cot(x)
tan_x = np.tan(x)
cot_x = 1 / np.tan(x)

# Create subplots
# fig : Figure object acts as a container
# axs : Array of Axes objects(actual plots)
# plt.subplots(rows, cols, figsize=(width, height))
fig, axs = plt.subplots(1, 3, figsize=(12, 6))

# Plot tan(x) on the first subplot
axs[0].plot(x, tan_x, label='tan(x)', color='b')
axs[0].set_title('Plot of tan(x)')
axs[0].set_ylim(-100, 100)  # Limit y-axis for better visualization
axs[0].axhline(0, color='black',linewidth=0.5)
axs[0].axvline(0, color='black',linewidth=0.5)
axs[0].grid(True)

# Plot cot(x) on the second subplot
axs[1].plot(x, cot_x, label='cot(x)', color='r')
axs[1].set_title('Plot of cot(x)')
axs[1].set_ylim(-100, 100)  # Limit y-axis for better visualization
axs[1].axhline(0, color='black',linewidth=0.5)
axs[1].axvline(0, color='black',linewidth=0.5)
axs[1].grid(True)

# Plot tan(x) + cot(x) on the second subplot
axs[2].plot(x, tan_x+cot_x, label='tan(x) + cot(x)', color='g')
axs[2].set_title('Plot of tan(x) + cot(x)')
axs[2].set_ylim(-100, 100)  # Limit y-axis for better visualization
axs[2].axhline(0, color='black',linewidth=0.5)
axs[2].axvline(0, color='black',linewidth=0.5)
axs[2].grid(True)

# Show the plots
plt.tight_layout()
plt.show()
```
