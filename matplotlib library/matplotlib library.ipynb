### matplotlib library
- plotting library

```python
# plot sin(x) and cos(x) in matplotlib with grids and coordinate system x and y axis

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 4 * np.pi, 50)
y = np.sin(x)
z = np.cos(x)

plt.plot(x, y, label='sin(x)')
plt.plot(x, z, label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin(x) and cos(x)')
plt.grid()
plt.legend()
plt.show()
```
