### can solve system of equations
- Linear
- Non -linear both

- Exact soln : `sympy.solve()`
- Numeric Solution : `sympy.nsolve()`

```python
from sympy.abc import x,y
from sympy import abc
from sympy import solve, Eq, lambdify

eq1 = Eq(2*x + 3*y, 6) # 2x + 3y = 6
eq2 = Eq(3*x + 2*y, 7) # 3x + 2y = 7

sol = solve((eq1, eq2), (x, y))
print(sol)
```

```python
# lambdify syntax : lambdify((variables), expression, 'numpy')
eq1_np = lambdify((x, y), eq1.lhs - eq1.rhs, 'numpy')
eq2_np = lambdify((x, y), eq2.lhs - eq2.rhs, 'numpy')


import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)
Z1 = eq1_np(X, Y)
Z2 = eq2_np(X, Y)

plt.contour(X, Y, Z1, levels=[0], colors='r')
plt.contour(X, Y, Z2, levels=[0], colors='b')
plt.plot(sol[abc.x], sol[abc.y], 'ro', label='Solution Point')


plt.show()
```
