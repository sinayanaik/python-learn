## typing
typing module in the Python standard library. It provides type hints so tools (mypy/pyright/IDE) can catch bugs before you run code.

```python
# Context: serial config object you pass around in scripts
from dataclasses import dataclass, field
from typing import Optional

@dataclass
class SerialCfg:
    port: Optional[str] = None          # str or None
    baud: int = 115200
```

```python
# Example usage
cfg1 = SerialCfg(port="/dev/ttyUSB0", baud=57600)
cfg2 = SerialCfg(baud=9600)
cfg3 = SerialCfg()

print(cfg1)
print(cfg2)
print(cfg3)
```
